***Importing the necessary libraries***

```

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Reading the dataset

```
dataset=pd.read_csv('Churn_Modelling.csv')
```

```
dataset.head()
```

```
dataset.columns
```

## Selecting the dependent variable Y and the features as X

```
from sklearn.utils import shuffle
dataset=shuffle(dataset)

X=dataset.drop(['RowNumber','CustomerId','Surname','Exited'],axis=1).values
Y=dataset['Exited'].values
```

## LAbel Encoding the variables and One Hot encoding the categorical variable

```
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```
# label encoding the Gender feature
LabelEncoder_Gender=LabelEncoder()
X[:,2]=LabelEncoder_Gender.fit_transform(X[:,2])
```

```
# label encoding the Geography feature
LabelEncoder_Geography=LabelEncoder()
X[:,1]=LabelEncoder_Geography.fit_transform(X[:,1])
```

```
# one hot encoding the Geography feature as it is a categorical feature
onehotencoder=OneHotEncoder(categorical_features=[1])
X=onehotencoder.fit_transform(X).toarray()
X=X[:,1:]         # to avoid the dummy variable trap
```

## **Data Normalisation**
### Scaling the data by subtracting from the mean and dividing by the standard deviation,to get a 0 mean and 1 std dataset

```
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
```

```
X=scalar.fit_transform(X)
```

## Splitting the dataset into training and testing part

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=.2)
```

```
import keras
from keras.models import Sequential
from keras.layers import Dense,Input
from keras.layers import Dropout   # to handle regularisation
```

```
classifier_model= Sequential()

first_hidden_layer=Dense(output_dim=6, input_dim=11, activation='relu', init='uniform')

classifier_model.add(first_hidden_layer)
classifier_model.add(Dropout(rate=0))    
# rate=p% means p% of the neurons will not contribute in the learning to prevent overfitting
# rate can very from 0 to 1, ideally should be lesser than 0.5, else it will result underfitting


second_hidden_layer=Dense(output_dim=6, activation='relu', init='normal')

classifier_model.add(second_hidden_layer)
classifier_model.add(Dropout(rate=0)) 



output_layer=Dense(output_dim=1, activation='sigmoid', init='normal')
classifier_model.add(output_layer)


classifier_model.compile(
    optimizer='rmsprop', 
    loss='binary_crossentropy', 
    metrics=['accuracy'])



classifier_model.fit(x=x_train,
                    y=y_train,
                    batch_size=125,
                    nb_epoch=500
                     )
```

## Model Evaluation

```
y_pred=classifier_model.predict(x_test)

y_pred[y_pred>=.5]=1
y_pred[y_pred<.5]=0
```

```
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test , y_pred)

accuracy= (cm[0][0]+cm[1][1])/len(y_test)
print("Accuracy is : ",accuracy*100," %")
```

### Use our ANN model to predict if the customer with the following informations will leave the bank:

```
def dopredict(geography,creditScore,gender,age,tenure,balance,numbeOfProducts,haveCard,isActiveMember,estimatedSalary):
    isSpain=1.0 if geography=="Spain" else 0.0
    isGermany=1.0 if geography=="Germany" else 0.0
    gender= 1 if gender=="Male" else 0
    info=scalar.transform(np.array([[isGermany, isSpain, creditScore, gender, age, tenure, balance, numbeOfProducts, haveCard, isActiveMember, estimatedSalary]]))
    prediction=classifier_model.predict(info) >=0.5
    prediction=prediction[0][0]
    print(prediction)
    if prediction:
        print("Yes,There is a strong possibility that the user will leave the Bank")
    else:
        print("No, the user will not leave the Bank")
```
